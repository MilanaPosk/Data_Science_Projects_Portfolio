## Загрузка данных и изучение общей информации

```python
import pandas as pd
data = pd.read_csv('/datasets/data.csv')
display(data.head(20))
data.isna().mean()
data.info()
```

**Вывод** В данном файле 12 столбцов. Есть вещественные данные в "общий трудовой стаж в днях" и "ежемесячный доход". Так же есть отрицательные значения в столбе "общий трудовой стаж в днях" п. И количество записей по столбцу "ежемесячный доход"  и "общий трудовой стаж в днях" меньше чем во все датасете, это связано с присутствием значений NaN в этих столбцах.

## Предобработка данных

### Обработка пропусков

```python
data.isnull().sum(axis = 0) # получили количество NaN вхождений в каждую колонку можно с помощью метода df.isnull().sum и передали axis = 0 в методе sum, чтобы получить количество NaN вхождений в каждом столбце. 

data['days_employed'] = data.groupby('dob_years').transform(lambda x: x.fillna(x.median())) # чтобы обработать NaN в "общий трудовой стаж в днях", используем медианный показатель сгруппированный по столбцу "возраст клиента в годах". 

data['total_income'] = data.groupby('income_type').transform(lambda x: x.fillna(x.median())) # чтобы обработать NaN в "тип занятости", используем медианный показатель сгруппированный по столбцу "тип занятости". 

display(data.head(20))
data.info() # проверим информацию о датасете. Все пропуски ушли.
```

**Вывод** Причиной пропусков могут быть с не трудоустройством или неофициальным заработком.  Также есть некая зависимость между отрицательными значениями и типов занятости. Возможно тут закралась ошибка технического характера, просто шкала учета идет в обратную сторону.  Так же данный метод transform и lambda помог избавиться от аномальных отрицательных значений.

### Замена типа данных

```python
data['total_income']=data['total_income'].astype('int') # метод для замен данных .astype() очень удобен, когда нужно конкретный тип данных столбца в другой тип данных, в нашем случае из float в int. 
data['days_employed']=data['days_employed'].astype('int')

display(data.head(20))
data.info() # проверим теперь общие данные.
```

**Вывод** На данном этапе ушли все вещественные значения. Теперь значения необходимых нам столбцов целочисленные и они не помешают нашим расчётам.

### Обработка дубликатов

```python
data.duplicated().sum() # проверим наличные дубликатов по всем столбцам
```

```python
data['education'].value_counts() #анализирует столбцы методам .value_counts(), выбирая каждое уникальное значение и подсчитывая частоту его встречаемости.
data['education'] = data['education'].str.lower() # столбец "уровень образования клиента" приводим к нижнему регистру вызовом метода lower(), чтоб исключить такого рода дубликаты.

data['purpose'].value_counts() #анализирует столбцы методам .value_counts(), выбирая каждое уникальное значение и подсчитывая частоту его встречаемости.
```

**Вывод** Мы убедились, что в данных есть дубликаты, 71. По столбцу "уровень образования клиента" наглядно видно, что все дело в уровне регистра. После проведения их к одному виду, данные стали читаемы. Для обработки дубликатов по столбцу "цель получения кредита" необходимо примерить лемматизацию данных, т.к. в них явно видно, что их можно категорировать. По обоим столбцам причиной дубликатов явно является ошибка ввода при заполнении данных о человеке, человеческий фактор.

### Лемматизация

```python
from pymystem3 import Mystem
m = Mystem() # вызываем одну из библиотек с функцией лемматизации на русском языке
purpose_list = list(data['purpose'])
for i in purpose_list: # перебираем циклом все элементы листа, созданного из по столбцу purpose
    lemmas = ' '.join(m.lemmatize(i)) # для наглядности итоговый результат объединили вызовом метода join()
    print(lemmas) 
```

```python
from collections import Counter
Counter(purpose_list) # рассчитываем, сколько раз каждое слово употребляется в столбце "цель получения кредита".
```

**Вывод** После визуального анализа лемматизированных данных, выделим 4 основных категории. Т.к. здесь тоже ошибка ввода изначальных данных анкетирования.

### Категоризация данных

```python
def categories_purpose(row): # создаем свою функцию с переменной для перебора элементов столбца purpose с условиями содержания в них определённых частей слов
    purpose = row['purpose']
    if 'недвижим' in purpose:
        return 'недвижимость'
    if 'жиль' in purpose:
        return 'недвижимость'
    if 'свадьб' in purpose:
        return 'свадьба'
    if 'автомобил' in purpose:
        return 'автомобиль'
    return 'образование'
# данные категории были выбраны исходя из основной сути цели кредитования
data['purpose_id'] = data.apply(categories_purpose, axis=1) # создаем новый столбец, для собрания там отобранных данных по четырем категориям для каждой строки
data
```

```python
def categories_total_income(row):
    total_income = row['total_income']
    if total_income <= 100000:
        return 'низкий уровень'
    if total_income <= 700000:
        return 'ниже среднего'
    if total_income <= 1000000:
        return 'средний'
    return 'высокий'
# данные категории были выбраны исходя из минимального и максимального уровня дохода
data['total_income_id'] = data.apply(categories_total_income, axis=1)
data
```

```python
def categories_children(row):
    children = row['children']
    if children == 0:
        return 'нет детей'
    if children <= 3:
        return '1-3 ребенка'
    return 'многодетные'
# данные категории были выбраны исходя из минимального и максимального количества детей, -1 и 20.
data['children_id'] = data.apply(categories_children, axis=1)
data
```

**Вывод** На данном этапе были прокатегорированы три столбца. Данные категории были сохранен в новые столбцы. Эти данные помогут для дальнейшего анализа их зависимости от возврата кредита в срок.

## Анализ

- Есть ли зависимость между наличием детей и возвратом кредита в срок?

```python
data_pivot_children = data.pivot_table(index=['children_id'], values='debt', aggfunc='mean')# для получения сводной таблицы используем метод pivot_table(), где в качестве аргументов у нас будут index - категорийные столбцы, values - возврат кредита, aggfunc - среднее значение. 
data_pivot_children.sort_values(by='debt', ascending=False) # отсортируем данные по столбцу debt в порядке убывания
```

**Вывод** Для определения зависимостей сведем данные в сводные таблицы. При значениях в столбце "задолженность по возврату кредитов", где 1 -есть задолженность, а 0 - нет, видно, что категории людей, имеющих детей. имеют большую долю задолженности в отличии от тех, у кого детей нет.

- Есть ли зависимость между семейным положением и возвратом кредита в срок?

```python
data_pivot_family_status = data.pivot_table(index=['family_status'], values='debt', aggfunc='mean') 
data_pivot_family_status.sort_values(by='debt', ascending=False) 
```

**Вывод** По данным сводной таблицы видно, что "Не женат / не замужем" и "гражданский брак" имеют наиболее часто встречающуюся задорностью.

- Есть ли зависимость между уровнем дохода и возвратом кредита в срок?

```python
data_pivot_total_income = data.pivot_table(index=['total_income_id'], values='debt', aggfunc='mean')
data_pivot_total_income.sort_values(by='debt', ascending=False)
```

**Вывод** По данным сводной таблицы видно, что уровень дохода не сильно влияет на задолженность по кредиту, потому что по всем категориям, от самого низкого до высокого, нет сильной разницы в доле. Т.е. наличие задолженности по уплате кредита в срок не зависит от уровня дохода заемщиков. 

- Как разные цели кредита влияют на его возврат в срок?

```python
data_pivot_purpose = data.pivot_table(index=['purpose_id'], values='debt', aggfunc='mean')
data_pivot_purpose.sort_values(by='debt', ascending=False)
```

**Вывод** По данным сводной таблицы видно, что основная цель кредита влияет на задолженность по нему. На покупку автомобиля и траты на образование приходится большая доля людей, имеющих задорность. А на свадьбу и недвижимость меньшее количество людей имеют долг перед банком.
